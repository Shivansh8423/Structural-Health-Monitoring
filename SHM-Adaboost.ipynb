```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from io import StringIO
from sklearn.preprocessing import MinMaxScaler
import sklearn
# test regression dataset
from sklearn.datasets import make_regression#1
from sklearn.ensemble import AdaBoostRegressor#2
from sklearn.metrics import r2_score,mean_squared_error
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/SHM final dataset/finaldataset.csv', dtype='unicode')
dataset.head(n=3)
dataset.shape
```

```python
#normalisation
scaler=MinMaxScaler()
scaler.fit(dataset.iloc[:,5:6])
dataset["Ground truth"]=scaler.transform(dataset.iloc[:,5:6])
dataset
```

```python
#To fix x value
X=dataset.iloc[:,0:5].values
#print(X)
```

```python
#To fix Y value
y=dataset.iloc[:,5].values
# print(y)
```

```python
#To fix train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.25,random_state=20000)
```

```python
#adaboost algo
model=AdaBoostRegressor(n_estimators = 200, learning_rate =
1.0, random_state = 80,loss='linear')
model.fit(X_train,Y_train)
#build boosted regressor from training set
```

```python
#To prediction
Y_pred=model.predict(X_test)#predict regression value for X_test
model.score(X,y)#return the coefficient of determination R^2 of prediction
#r2=r2_score(Y_test,Y_pred)
#r2
#plt.scatter(dataset["Cycle"],dataset["Ground truth"])
#plt.title("adaboost")
```

```python
r2_score(Y_test,Y_pred)#r2_score(y_true, y_pred)
```

```python
Y_pred_x=model.predict(X_train)
r2_score(Y_train,Y_pred_x)
```

```python
#Formula for calculating RMSE
mse=mean_squared_error(Y_test, Y_pred)
print(mse)
rmse=np.sqrt(mse)
rmse
#mean_squared_error(Y_test,Y_pred)
```
